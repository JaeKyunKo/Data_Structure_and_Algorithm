```
class Node :
  value, nodeNext = None, None
  blnHead, blnTail = False, False
  def __init__(self, value = None, nodeNext = None, blnHead = False, blnTail = False) :
    self.value = value
    self.nodeNext = nodeNext
    self.blnHead = blnHead
    self.blnTail = blnTail
  def getValue(self) :
    return self.value
  def getNext(self) :
    return self.nodeNext
  def setValue(self, vlaue) :
    self.value = value
  def setNext(self, nodeNext) :
    self.nodeNext = nodeNext
  def isHead(self) :
    return self.blnHead
  def isTail(self) :
    return self.blnTail
```

```
class SinglyLinkedList :
  nodeHead, nodeTail = None, None
  size = 0
  def __init__(self) :
    self.nodeTail = Node(blnTail = True)
    self.nodeHead = Node(blnHead = True, nodeNext = self.nodeTail)
  def getNode(self, idxRetrieve) :
    nodeReturn = self.nodeHead
    for i in range(idxRetrieve + 1) :
      nodeReturn = nodeReturn.getNext()
    return nodeReturn
  def getSize(self) :
    return self.size
  def printStatus(self) :
    nodeCurrent = self.nodeHead
    while nodeCurrent.getNext().isTail() is False :
      nodeCurrent = nodeCurrent.getNext()
      print(nodeCurrent.getValue(), end = " ")
    print()
  def insertAt(self, value, idxInsert) :
    nodePrev = self.getNode(idxInsert - 1)
    nodeNext = nodePrev.getNext()
    nodeInsert = Node(value, nodeNext = nodeNext)
    nodePrev.setNext(nodeInsert)
    self.size += 1
    return
  def removeAt(self, idxRemove) :
    nodePrev = self.getNode(idxRemove - 1) 
    nodeRemove = nodePrev.getNext()
    nodeNext = nodeRemove.getNext()
    nodePrev.setNext(nodeNext)
    self.size -= 1
    return nodeRemove.getValue()
```

```
class Queue(object) :
  queue = None
  def __init__(self) :
    self.queue = SinglyLinkedList()
  def enqueue(self, value) :
    self.queue.insertAt(value, 0)
  def dequeue(self) :
    return self.queue.removeAt(self.queue.getSize() - 1)
  def isEmpty(self) :
    if self.queue.getSize() == 0 :
      return True
    else :
      return False
  def printStatus(self) :
    self.queue.printStatus()
```

```
class TreeNode :
  value, nodeParent, nodeLHS, nodeRHS = None, None, None, None
  def __init__(self, value, nodeParent) :
    self.value = value
    self.nodeParent = nodeParent
  def getValue(self) :
    return self.value
  def getParent(self) :
    return self.nodeParent
  def getLHS(self) :
    return self.nodeLHS
  def getRHS(self) :
    return self.nodeRHS
  def setValue(self, value) :
    self.value = value
  def setParent(self, nodeParent) :
    self.nodeParent = nodeParent
  def setLHS(self, nodeLHS) :
    self.nodeLHS = nodeLHS
  def setRHS(self, nodeRHS) :
    self.nodeRHS = nodeRHS
```

```
class BinarySearchTree :
  root = None
  def __init__(self) :
    pass
  def insert(self, value, node = None) :
    if node is None :
      node = self.root
    if self.root is None :
      self.root = TreeNode(value, None)
      return
    if value == node.getValue() :
      return
    elif value > node.getValue() :
      if node.getRHS() is None :
        node.setRHS(TreeNode(value, node))
      else :
        self.insert(value, node.getRHS())
    else :
      if node.getLHS() is None :
        node.setLHS(TreeNode(value, node))
      else :
        self.insert(value, node.getLHS())
    return
  def search(self, value, node = None) :
    if node is None :
      node = self.root
    if value == node.getValue() :
      return True
    elif value > node.getValue() :
      if node.getRHS() is None :
        return False
      else :
        return self.search(value, node.getRHS())
    else :
      if node.getLHS() is None :
        return False
      else :
        return self.search(value, node.getLHS())
  def delete(self, value, node = None) :
    if node is None :
      node = self.root
    if value > node.getValue() :
      return self.delete(value, node.getRHS())
    elif value < node.getValue() :
      return self.delete(value, node.getLHS())
    else :
      if node.getLHS() is not None and node.getRHS() is not None :
        nodeReplace = self.findMin(node.getRHS())
        node.setValue(nodeReplace)
        self.delete(nodeReplace, node.getRHS())
        return
      parent = node.getParent()
      if node.getLHS() is not None :
        if node == self.root :
          self.root = node.getLHS()
        elif parent.getLHS() == node :
          parent.setLHS(node.getLHS())
          node.getLHS().setParent(parent)
        else :
          parent.setRHS(node.getLHS())
          node.getLHS().setParent(parent)
        return
      if node.getRHS() is not None :
        if node == self.root :
          self.root = node.getRHS()
        elif parent.getRHS() == node :
          parent.setRHS(node.getRHS())
          node.getRHS().setParent(parent)
        else :
          parent.setLHS(node.getRHS())
          node.getRHS().setParent(parent)
        return
    if node == self.root :
      self.root = None
    elif parent.getLHS() == node :
      parent.setLHS(None)
    else :
      parent.setRHS(None)
  def findMax(self, node = None) :
    if node is None :
      node = self.root
    if node.getRHS() is None :
      return node.getValue()
    else :
      return self.findMax(node.getRHS())
  def findMin(self, node = None) :
    if node is None :
      node = self.root
    if node.getLHS() is None :
      return node.getValue()
    else :
      return self.findMin(node.getLHS())
  def traversePreOrder(self, node = None) :
    if node is None :
      node = self.root
    returnList = list()
    returnList.append(node.getValue())
    if node.getLHS() is not None :
      returnList += self.traversePreOrder(node.getLHS())
    if node.getRHS() is not None :
      returnList += self.traversePreOrder(node.getRHS())
    return returnList
  def traverseInOrder(self, node = None) :
    if node is None :
      node = self.root
    returnList = list()
    if node.getLHS() is not None :
      returnList += self.traverseInOrder(node.getLHS())
    returnList.append(node.getValue())
    if node.getRHS() is not None :
      returnList += self.traverseInOrder(node.getRHS())
    return returnList
  def traversePostOrder(self, node = None) :
    if node is None :
      node = self.root
    returnList = list()
    if node.getLHS() is not None :
      returnList += self.traversePostOrder(node.getLHS())
    if node.getRHS() is not None :
      returnList += self.traversePostOrder(node.getRHS())
    returnList.append(node.getValue())
    return returnList
  def traverseLevelOrder(self) :
    returnList = list()
    q = Queue()
    q.enqueue(self.root)
    while q.isEmpty() is False :
      nodeRemove = q.dequeue()
      if nodeRemove is None :
        continue
      returnList.append(nodeRemove.getValue())
      if nodeRemove.getLHS() is not None :
        q.enqueue(nodeRemove.getLHS())
      if nodeRemove.getRHS() is not None :
        q.enqueue(nodeRemove.getRHS())
    return returnList
```
